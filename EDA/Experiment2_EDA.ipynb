## Learning data

### learning_data.csv

- subject_id: Unique identifier for each participant in the experiment.
- block_condition: Indicates whether the block was category-predictive (where all members of a category are rewarded for the same choice) or exemplar-predictive (where the optimal choice differs between exemplars within a category).
- block_order: The order in which the block was presented to the participant (e.g., 1st block, 2nd block, etc.).
- task_block: The same as block_condition, but could potentially refer to a broader task category (e.g., learning vs memory test).
- learning_trial: Identifies a specific trial within a learning block.
- learning_trial_group: Could indicate if the trial belongs to a specific group within the learning block (e.g., first half, second half).
- stimulus: The image presented to the participant during the trial.
- stimulus_repetition: Tracks how many times a particular stimulus has been presented throughout the experiment.
- left_choice, middle_choice, right_choice: The text or description of the choices presented on the left, middle, and right side of the screen during the trial.
- key_press: The key the participant pressed (1, 2, or 3) to indicate their choice.
- option_selected: Likely the numerical value (1, 2, or 3) corresponding to the choice the participant selected.
- rt: Reaction time - the time it took the participant to make their choice.
- points: The number of points earned by the participant on that trial (positive for correct, negative for incorrect).
- correct: Indicates whether the participant's choice was correct (1) or incorrect (0).
- optimal_choice: The numerical value (1, 2, or 3) corresponding to the choice that would have resulted in the most points for that trial.

### Data Visualisation

```python
# Histogram of 'rt'
plt.figure(figsize=(8, 6))
sns.histplot(data=learning_e2, x='rt', bins=20, kde=True)
plt.title('Distribution of Reaction Time (rt)')
plt.xlabel('Reaction Time (rt)')
plt.ylabel('Frequency')
plt.show()
```

#### Inferences

- Distribution: The distribution of reaction times is likely right-skewed. This means that there are more trials with faster reaction times than there are trials with slower reaction times.
- Center: It appears to be around 1000 ms.
- Variability: The spread of the reaction times appears to be large. There are trials with reaction times as fast as 200 ms and some as slow as 4000 ms.


```python
# Group by subject_id and aggregate the desired metrics
subject_performace = learning_e2.groupby('subject_id').agg({
    'points': 'sum',  # Total points scored by each subject
    'correct': lambda x: (x == 1).sum(),  # Number of correct responses
    'rt': 'mean'  # Mean reaction time
}).reset_index()
```

```python
subject_performace
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=subject_performace, x='age', y='rt', palette='crest')
plt.title('Age vs Mean Reaction Time')
plt.xlabel('Age')
plt.ylabel('Mean Reaction Time')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=subject_performace, x='age', y='points', palette='crest')
plt.title('Age vs Total Points')
plt.xlabel('Age')
plt.ylabel('Total Points')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.pairplot(subject_performace[['age', 'points', 'correct', 'rt']])
plt.suptitle('Pairplot of Age, Points, Correct Responses, and Mean Reaction Time')
plt.tight_layout()
plt.show()
```

```python
# Scatter plot of 'rt' vs 'points'
plt.figure(figsize=(10, 6))
sns.scatterplot(data=learning_e2, x='rt', y='points', hue='correct')
plt.title('Reaction Time vs Points (Colored by Correctness)')
plt.xlabel('Reaction Time (ms)')
plt.ylabel('Points')
plt.legend(title='Correct')
plt.show()
```

#### Inferences

- Faster reaction times are associated with higher points. The data points are concentrated in the lower left corner of the graph, indicating that faster reaction times (on the x-axis) are associated with earning more points (represented by the color gradient, with darker blue indicating higher points).

- There is a spread of reaction times for both correct and incorrect responses. While the trend suggests faster times lead to more points, there is still a range of reaction times for both correct and incorrect responses. For instance, there are some slow responses (on the right side of the x-axis) that are still colored blue, which means they were correct. Conversely, there are some fast responses (on the left side of the x-axis) that are colored orange, meaning they were incorrect.


```python
plt.figure(figsize=(10, 6))
sns.countplot(data=learning_e2, x='points', palette='crest')
plt.title('Distribution of Points', fontsize=16)  
plt.xlabel('Points', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Count plot of 'stimulus_repetition' colored by 'correct'
plt.figure(figsize=(10, 6))
sns.countplot(data=learning_e2, x='stimulus_repetition', hue='correct')
plt.title('Count of Correct Responses by Stimulus Repetition')
plt.xlabel('Stimulus Repetition')
plt.ylabel('Count')
plt.legend(title='Correct')
plt.show()
```

#### Inferences

```python
# correlation matrix
learning_numeric = learning_e2.select_dtypes(include='number')
correlation_matrix = learning_numeric.corr()
mask = np.tril(np.ones_like(correlation_matrix, dtype=bool))
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix for Learning Data')
plt.show()
```

```python
# Merge with age data
merged_data = pd.merge(learning_e2, sub_e2, on='subject_id')

# Round ages for smoother visualization
merged_data['rounded_age'] = np.round(merged_data['age'])

# Calculate average reaction time by rounded age
avg_rt_by_rounded_age = merged_data.groupby('rounded_age')['rt'].mean().reset_index()
plt.figure(figsize=(10, 6))
sns.lineplot(data=avg_rt_by_rounded_age, x='rounded_age', y='rt', color='blue', alpha=0.5, marker='o')
plt.title('Average Reaction Time by Rounded Age', fontsize=16)
plt.xlabel('Age', fontsize=14)
plt.ylabel('Average Reaction Time', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

#### Inferences

- On average, reaction time slows with age, indicated by the upward slope of the line as age increases. Individuals in their early teens (around 12.5 years old) exhibit the fastest reaction times, approximately 1000 milliseconds. Reaction time then gradually increases until around 25 years old, stabilizing at approximately 1300 milliseconds.
- Individual variation exists around this average trend, and other factors like health, fitness, and fatigue can influence reaction time.
- The rate of increase in reaction time diminishes with age, suggesting that older individuals experience a slower rate of change in reaction time compared to younger individuals.
- Despite age-related changes, the difference in reaction time between young and older individuals is relatively modest.


```python
# Calculate average points by age
avg_points_by_age = merged_data.groupby('age')['points'].mean().reset_index()

plt.figure(figsize=(10, 6))
avg_points_by_rounded_age = merged_data.groupby('rounded_age')['points'].mean().reset_index()
sns.lineplot(data=avg_points_by_rounded_age, x='rounded_age', y='points', color='blue', alpha=0.5, marker='o')
plt.title('Average Points by Rounded Age', fontsize=16)
plt.xlabel('Age', fontsize=14)
plt.ylabel('Average Points', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()  
plt.show()
```

#### Inferences

- We can infer that performance on this test increases with age. This is because the line slopes upwards as age increases.

## Memory data

### memory_data.csv
- subject_id: Unique identifier for each participant in the experiment.
- task_part: Indicates whether this data is from the learning task or the memory test.
- block_order: The order in which the block was presented to the participant during the learning task (e.g., 1st block, 2nd block, etc.).
- block_condition: Indicates whether the corresponding block in the learning task was category-predictive (all members of a category are rewarded for the same choice) or exemplar-predictive (optimal choice differs between exemplars within a category).
- category_type: The category that the image belongs to ie old or new.
- image: The name of the image file presented during the memory test.
- category: The category that the image belongs to.
- task_block: Potentially refers to a specific block within the memory test itself.
- stim_num: Stimulus number within the block (might correspond to presentation order).
- correct_mem_response: Indicates whether the participant's response (old vs new) to the image was correct (1) or incorrect (0) during the memory test.
- mem_response: The participant's response (old vs new) to the image during the memory test.
- mem_response_confidence: The participant's reported confidence level in their memory test response (e.g., definitely old, probably old).
- tmem_accconf_number: Likely a numerical code corresponding to the confidence level (e.g., 1=definitely old, 2=probably old).
- conf_rating: The participant's confidence rating on a scale (possibly numerical).
- foil_type: Indicates whether the image was an old image seen during learning (old) or a new image not seen before (foil).
- memory_delay: The number of days between the learning task and the memory test.

```python
#Histogram of Memory Response Time (rt):
plt.figure(figsize=(8, 6))
sns.histplot(data=mem_e2, x='rt', bins=20, kde=True, color='skyblue')
plt.title('Distribution of Memory Response Time')
plt.xlabel('Response Time (ms)')
plt.ylabel('Frequency')
plt.show()
```

#### Inferences

- Most frequent response times: The majority of memory response times fall within the range of 1000 to 3000 milliseconds. This suggests that the typical time taken to retrieve data from memory is between 1 and 3 seconds.

```python
# Merge memory data with subject ages
merged_data = pd.merge(mem_e2, sub_e2, on='subject_id')

# Calculate the proportion of correct memory responses for each age group
merged_data['rounded_age'] = np.round(merged_data['age'])
mem_acc_by_age = merged_data.groupby('rounded_age')['mem_acc'].mean().reset_index()

# Plot memory accuracy by age using a line plot
plt.figure(figsize=(12, 6))
sns.lineplot(data=mem_acc_by_age, x='rounded_age', y='mem_acc', color='blue')
plt.title('Memory Accuracy by Age', fontsize=16)
plt.xlabel('Age', fontsize=14)
plt.ylabel('Proportion of Correct Memory Responses', fontsize=14)
plt.xticks(fontsize=12, rotation=45)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

#### Inferences

- The graph shows a decline in memory accuracy as age increases.
- This suggests memory weakens as we age.
- Possible reasons include difficulty encoding new information and retrieving stored memories.
- Individual variability exists; some older adults have sharp memory.

```python
# Calculate the average memory delay for each age group
mem_delay_by_age = merged_data.groupby('rounded_age')['memory_delay'].mean().reset_index()

# Plot memory delay by age using a bar plot
plt.figure(figsize=(12, 6))
sns.barplot(data=mem_delay_by_age, x='rounded_age', y='memory_delay', palette='crest')
plt.title('Memory Delay by Age', fontsize=16)
plt.xlabel('Age', fontsize=14)
plt.ylabel('Memory Delay', fontsize=14)
plt.xticks(fontsize=12, rotation=45)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

## RL Data

### rl_data.csv

- subject_id: Unique identifier for each participant in the experiment.
- block_order: The order in which the block was presented to the participant (e.g., 1st block, 2nd block, etc.).
- block_condition: Indicates whether the block was category-predictive (all members of a category are rewarded for the same choice) or exemplar-predictive (optimal choice differs between exemplars within a category).
- trial: Identifies a specific trial within a learning block.
- cat_num: Category number that the stimulus belongs to (e.g., 1 might represent "vehicles").
- exemplar_num: Exemplar number within a category (likely 1 to however many exemplars were used per category).
- correct_choice_num: The numerical value (1, 2, or 3) corresponding to the choice that would have resulted in the most points for that trial.
- left_choice_num, middle_choice_num, right_choice_num: The numerical values (1, 2, or 3) corresponding to the choices presented on the left, middle, and right side of the screen during the trial.
- choice: The participant's chosen option during the trial (numerical value: 1, 2, or 3).
- choice_position: The position (left, middle, or right) where the participant's chosen option was presented on the screen.
- points: The number of points earned by the participant on that trial (positive for correct, negative for incorrect).
