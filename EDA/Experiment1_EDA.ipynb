### Data Visualization

```python
# Distribution of ages using histogram
plt.figure(figsize=(10, 6))
sns.histplot(e1_sub_ages['age'], kde=True, bins=20, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Distribution of Subject Ages', fontsize=16)  
plt.xlabel('Age', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

#### Inferences

The x-axis represents age, ranging from 7.5 to 25 years old, with increments of 2.5 years. The y-axis represents frequency, which indicates the number of subjects in each age group.

From the plot, we can infer the following:

- The most frequent age group is between 10 and 12.5 years old. This age group contains more subjects than any other.
- The number of subjects generally decreases as age increases. There are notably fewer subjects over 20 years old.
- The distribution appears somewhat symmetrical, with a similar number of subjects in the age groups below 12.5 years old as there are in the age groups above 12.5 years old.


```python
# Boxplot to see the distribution of ages
plt.figure(figsize=(8, 5))
sns.boxplot(data=e1_sub_ages, x='age', color='skyblue', width=0.3, linewidth=1.5, fliersize=5)
plt.title('Boxplot of Subject Ages', fontsize=16)
plt.xlabel('Age', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

#### Inferences

- The middle 50% of subjects have ages between 10 and 17.5 years old. This range is represented by the box in the plot, known as the interquartile range (IQR).
- The median, shown as a line inside the box, divides the data into two halves, with half the values above and half below.
- Data spread is wider for ages younger than 10 than for ages older than 17.5, as indicated by the length of the whiskers.
- Whiskers extend to the most extreme values within 1.5 times the IQR from the bottom and top of the box.
- The longer whisker below the box suggests a wider range of values for younger ages.
- There may be outliers, indicated by points beyond the whiskers. Two circles plotted above the whisker could be outliers.

```python
# Calculate mean reaction times for Category Predictive condition
mean_rt_category = e1_learning_category.groupby('stimulus_repetition')['rt'].mean().reset_index()

# Calculate the standard error of the mean for reaction times in Category Predictive condition
sem_rt_category = e1_learning_category.groupby('stimulus_repetition')['rt'].sem().reset_index()

# Calculate mean reaction times for Exemplar Predictive condition
mean_rt_exemplar = e1_learning_exemplar.groupby('stimulus_repetition')['rt'].mean().reset_index()

# Calculate the standard error of the mean for reaction times in Exemplar Predictive condition
sem_rt_exemplar = e1_learning_exemplar.groupby('stimulus_repetition')['rt'].sem().reset_index()

# Plotting the line plot with error bars for both conditions
plt.figure(figsize=(10, 6))

# Plot Category Predictive condition
sns.lineplot(x='stimulus_repetition', y='rt', data=mean_rt_category, 
             marker='o', err_style='bars', ci='sd', label='Category Predictive')

# Plot Exemplar Predictive condition
sns.lineplot(x='stimulus_repetition', y='rt', data=mean_rt_exemplar, 
             marker='o', err_style='bars', ci='sd', label='Exemplar Predictive')

plt.title('Stimulus Repetition vs Mean Reaction Times with Error Bars')
plt.xlabel('Stimulus Repetition')
plt.ylabel('Mean Reaction Times')
plt.legend()
plt.grid(True)
plt.show()
```

- The decline in reaction time appears to be steeper than the decline in points. 
- The slope of the line for average reaction time is steeper than the slope of the line for average points.

### rl_data.csv

1. **subject_id:** Unique identifier for each participant.
   
2. **block_order:** Indicates the order in which the blocks of the experiment were presented to the participant.

3. **block_condition:** Represents the condition of the block (category-predictive or exemplar-predictive).

4. **trial:** Unique identifier for each trial, distinguishing individual instances throughout the experiment.

5. **cat_number:** Indicates the category number of the stimulus presented in the trial.

6. **stim_number:** Represents the specific stimulus number within a category for each trial.

7. **approach_reward:** The reward associated with choosing to "approach" the stimulus. This value influences the reinforcement learning process.

8. **choice:** Indicates whether the participant chose to "approach" (1) or "avoid" (2) the stimulus, reflecting their decision-making during the trial.

9. **points:** The points earned or lost based on the participant's choice and the associated reward. This contributes to the reinforcement aspect of the learning task.
