## Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

## Subject Age

### Loading Data

```python
# reading subject age data
e1_sub_ages = pd.read_csv('./data/e1/e1_sub_ages.csv')
```

```python
e1_sub_ages.head()
```

### Data Description

```python
e1_sub_ages.info()
```

```python
e1_sub_ages.describe()
```

- There are 151 subjects.
- The age range is `8-25` years old.
- The average age is `15.8` years old.

### Data Visualization

```python
# Distribution of ages using histogram
plt.figure(figsize=(10, 6))
sns.histplot(e1_sub_ages['age'], kde=True, bins=20, color='orange', edgecolor='black', alpha=0.7)
plt.title('Distribution of Subject Ages', fontsize=16)  
plt.xlabel('Age', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Boxplot to see the distribution of ages
plt.figure(figsize=(8, 5))
sns.boxplot(data=e1_sub_ages, x='age', color='blue', width=0.3, linewidth=1.5, fliersize=5)
plt.title('Boxplot of Subject Ages', fontsize=16)
plt.xlabel('Age', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

## Learning Data

### Loading Data

```python
# reading learning data
e1_learning = pd.read_csv('./data/e1/e1_learning_data.csv')
e1_learning.head()
```

### Data Description

```python
e1_learning.info()
```

```python
e1_learning.describe()
```

### Data Visualization

```python
e1_learning['mean_cat_reward'].value_counts()
```

```python
# Histogram of 'mean_cat_reward'
plt.figure(figsize=(10, 6))
sns.histplot(e1_learning['mean_cat_reward'], bins=20, color='green', edgecolor='black', alpha=0.7)
plt.title('Distribution of Mean Category Reward', fontsize=16)  
plt.xlabel('Mean Category Reward', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
e1_learning['approach_reward'].value_counts().sort_index().plot(kind='bar', color = 'blue', edgecolor='black', alpha=0.7)
plt.title('Approach Reward', fontsize=16)
plt.xlabel('Approach Reward', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Histogram of 'rt'
plt.figure(figsize=(10, 6))
sns.histplot(e1_learning['rt'], kde=True, bins=20, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Distribution of Reaction Time (rt)', fontsize=16)  
plt.xlabel('Reaction Time (rt)', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=e1_learning, x='points', hue = 'points', palette='pastel', legend=False)
plt.title('Distribution of Points', fontsize=16)  
plt.xlabel('Points', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Box plot of 'mean_cat_reward' by 'cat_type'
plt.figure(figsize=(10, 6))
sns.boxplot(data=e1_learning, x='cat_type', y='mean_cat_reward', palette='pastel', linewidth=1.5, fliersize=5, hue = 'cat_type')
plt.title('Mean Category Reward by Category Type', fontsize=16)
plt.xlabel('Category Type', fontsize=14)
plt.ylabel('Mean Category Reward', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Scatter plot of 'rt' vs 'stimulus_repetition' using hexbin
plt.figure(figsize=(10, 8))
sns.jointplot(data=e1_learning, x='rt', y='stimulus_repetition', kind='hex', color='skyblue')
plt.title('Hexbin Plot of Reaction Time vs Stimulus Repetition', pad=100)
plt.xlabel('Reaction Time', labelpad=20)
plt.ylabel('Stimulus Repetition', labelpad=20)
plt.show()
```

```python

e1_learning['approach_reward'] = e1_learning['approach_reward'].str.replace('[', '').str.replace(']', '').astype(float)
```

```python
# One-hot encode categorical variables
e1_learning_encoded = pd.get_dummies(e1_learning, columns=['task_block', 'cat_type'])

e1_learning_corr = e1_learning_encoded.drop(['subject_id', 'task_part', 'stimulus', 'key_press'], axis=1)

corr_matrix = e1_learning_corr.corr()

# Heatmap of correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=e1_learning, x='mean_cat_reward', y='approach_reward', color='skyblue', alpha=0.5)
plt.title('Scatter Plot of Mean Cat Reward vs Approach Reward', fontsize=16)
plt.xlabel('Mean Cat Reward', fontsize=14)
plt.ylabel('Approach Reward', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
aggregated_data = e1_learning.groupby('mean_cat_reward')['points'].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.lineplot(data=aggregated_data, x='mean_cat_reward', y='points', color='red', alpha=0.5)
plt.title('Line Plot of Mean Cat Reward vs Points (Aggregated)', fontsize=16)
plt.xlabel('Mean Cat Reward', fontsize=14)
plt.ylabel('Mean Points', fontsize=14)  
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=e1_learning, x='approach_reward', y='points', color='violet', alpha=0.5)
plt.title('Scatter Plot of Approach Reward vs Points', fontsize=16)
plt.xlabel('Approach Reward', fontsize=14)
plt.ylabel('Points', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
e1_learning['correct_response'].value_counts()
```

```python
# Pairplot for multivariate analysis
sns.pairplot(e1_learning[['rt', 'stimulus_repetition','correct_response']])
plt.title('Pairplot of Numeric Variables')
plt.show()
```

```python
merged_data = pd.merge(e1_learning, e1_sub_ages, on='subject_id')
```

```python
avg_rt_by_age = merged_data.groupby('age')['rt'].mean().reset_index()

plt.figure(figsize=(15, 6))
plt.subplot(1, 2, 1)
sns.lineplot(data=avg_rt_by_age, x='age', y='rt', color='blue', alpha=0.5, marker='o')
plt.title('Average Reaction Time by Age', fontsize=16)
plt.xlabel('Age', fontsize=14)
plt.ylabel('Average Reaction Time', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)

# Plot with rounded ages
merged_data['rounded_age'] = np.round(merged_data['age'])
avg_rt_by_rounded_age = merged_data.groupby('rounded_age')['rt'].mean().reset_index()

plt.subplot(1, 2, 2)
sns.lineplot(data=avg_rt_by_rounded_age, x='rounded_age', y='rt', color='blue', alpha=0.5, marker='o')
plt.title('Average Reaction Time by Rounded Age', fontsize=16)
plt.xlabel('Age', fontsize=14)
plt.ylabel('Average Reaction Time', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout() 
plt.show()
```

```python
avg_points_by_age = merged_data.groupby('age')['points'].mean().reset_index()

plt.figure(figsize=(15, 6))
plt.subplot(1, 2, 1)
sns.lineplot(data=avg_points_by_age, x='age', y='points', color='green', alpha=0.5, marker='o')
plt.title('Average points by Age', fontsize=16)
plt.xlabel('Age', fontsize=14)
plt.ylabel('Average points', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)

# Plot with rounded ages
merged_data['rounded_age'] = np.round(merged_data['age'])
avg_rt_by_rounded_age = merged_data.groupby('rounded_age')['points'].mean().reset_index()

plt.subplot(1, 2, 2)
sns.lineplot(data=avg_rt_by_rounded_age, x='rounded_age', y='points', color='green', alpha=0.5, marker='o')
plt.title('Average points by Rounded Age', fontsize=16)
plt.xlabel('Age', fontsize=14)
plt.ylabel('Average Points', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()  
plt.show()
```
